DATA CLEANING

```python
#Imports
import pandas as pd
```

```python
#Load Data
heart_disease_df = pd.read_csv('Resources/heart_disease_dataset.csv')
heart_disease_df.head()
```

```python
heart_disease_df = heart_disease_df.drop('Unnamed: 0.1', axis = 1)
heart_disease_df = heart_disease_df.drop('Unnamed: 0', axis = 1)
```

```python
heart_disease_df.head()
```

```python
heart_disease_df.isnull().sum()
```

```python
heart_disease_df['Alcohol Intake'] = heart_disease_df['Alcohol Intake'].fillna("None")
heart_disease_df.head()
```

```python
heart_disease_matrix = heart_disease_df
heart_disease_matrix
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
heart_disease_matrix['Gender'] = le.fit_transform(heart_disease_matrix['Gender'])
heart_disease_matrix['Smoking'] = le.fit_transform(heart_disease_matrix['Smoking'])
heart_disease_matrix['Alcohol Intake'] = le.fit_transform(heart_disease_matrix['Alcohol Intake'])
heart_disease_matrix['Family History'] = le.fit_transform(heart_disease_matrix['Family History'])
heart_disease_matrix['Diabetes'] = le.fit_transform(heart_disease_matrix['Diabetes'])
heart_disease_matrix['Obesity'] =le.fit_transform(heart_disease_matrix['Obesity'])
heart_disease_matrix['Exercise Induced Angina']=le.fit_transform(heart_disease_matrix['Exercise Induced Angina'])
heart_disease_matrix['Chest Pain Type']=le.fit_transform(heart_disease_matrix['Chest Pain Type'])
heart_disease_matrix.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
heart_disease_numeric = heart_disease_matrix.select_dtypes(include=['int64'])
correlation_matrix = heart_disease_numeric.corr()['Heart Disease'].drop('Heart Disease')

#plt.figure(figsize=(10, 8))  # Adjust the size as neede
correlation_matrix.plot(kind="bar")
plt.title('Correlation Matrix of Heart Disease Dataset')
plt.show()
plt.savefig('Images/corr_matrix.png')
```
