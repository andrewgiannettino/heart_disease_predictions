DATA CLEANING

```python
#Imports
import pandas as pd
```

```python
#Load Data
heart_disease_df = pd.read_csv('Resources/heart_disease_dataset.csv')
heart_disease_df.head()
```

```python
heart_disease_df = heart_disease_df.drop('Unnamed: 0.1', axis = 1)
heart_disease_df = heart_disease_df.drop('Unnamed: 0', axis = 1)
```

```python
heart_disease_df.head()
```

```python
heart_disease_df.isnull().sum()
```

```python
heart_disease_df['Alcohol Intake'] = heart_disease_df['Alcohol Intake'].fillna("None")
heart_disease_df.head()
```

```python
heart_disease_matrix = heart_disease_df
heart_disease_matrix
```

```python
#Turning Gender Column into Int
def replace_words(word):
    if word == 'Female':
        return 0
    elif word == 'Male':
        return 1
    else:
        return word
heart_disease_matrix['Gender'] = heart_disease_matrix['Gender'].apply(replace_words)
heart_disease_matrix.head()
```

```python
#Turning smoking columbn into int values
def replace_words(word):
    if word == 'Never':
        return 0
    elif word == 'Current':
        return 1
    elif word == 'Former':
        return 3
    else:
        return word
heart_disease_matrix['Smoking'] = heart_disease_matrix['Smoking'].apply(replace_words)
heart_disease_matrix.head()
```

```python
#Turning Alcohol Intake columbn into int values
def replace_words(word):
    if word == 'None':
        return 0
    elif word == 'Moderate':
        return 1
    elif word == 'Heavy':
        return 3
    else:
        return word
heart_disease_matrix['Alcohol Intake'] = heart_disease_matrix['Alcohol Intake'].apply(replace_words)
heart_disease_matrix.head()
```

```python
#Turning Family History column into int values
def replace_words(word):
    if word == 'No':
        return 0
    elif word == 'Yes':
        return 1
    else:
        return word
heart_disease_matrix['Family History'] = heart_disease_matrix['Family History'].apply(replace_words)
heart_disease_matrix.head()
```

```python
#Turning Diabetes column into int values
def replace_words(word):
    if word == 'No':
        return 0
    elif word == 'Yes':
        return 1
    else:
        return word
heart_disease_matrix['Diabetes'] = heart_disease_matrix['Diabetes'].apply(replace_words)
heart_disease_matrix.head()
```

```python
#Turning Obesity column into int values
def replace_words(word):
    if word == 'No':
        return 0
    elif word == 'Yes':
        return 1
    else:
        return word
heart_disease_matrix['Obesity'] = heart_disease_matrix['Obesity'].apply(replace_words)
heart_disease_matrix.head()
```
